1. Read csv file of Canada

```python
df = pd.read_csv('Canada.csv')
df.head()
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df['Borough'].unique()),
        df.shape[0]
    )
)
```

Here i have clustered only the neighborhoods in Noth York, I sliced the original dataframe and created a new dataframe of the North York data.

```python
north_york = df[df['Borough']=='North York'].reset_index(drop=True)
north_york.head()
```

```python
address = 'North York'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of North York City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_northyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_csv=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_csv=False).add_to(map_northyork)  
    
map_northyork
```

Folium is a great visualization library, you can zoom into the above map, and click on each circle mark to reveal the name of the neighborhood and its respective borough.
